```python
train = pd.read_csv('/content/drive/MyDrive/WOC 4 (2).0/Polynomial_train.csv')
train.head()
```

```python
normalized_dataset = (train - np.min(train)) / (np.max(train) - np.min(train))
X = normalized_dataset.iloc[:,1:4]
Y = normalized_dataset.iloc[:,4]

X_train=X.to_numpy()
Y_train=Y.to_numpy()

m=Y_train.shape[0]
X_train=np.c_[np.ones(m),X_train]
Y_train= Y_train.reshape(X_train.shape[0],1)

print(X_train.shape)
print(Y_train.shape)
```

```python
X=X_train.astype('float') 
Y=Y_train.astype('float')
```

```python
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        X[i][j]=X[i][j]**(j+1)

print(X)

theta = np.matmul(np.linalg.inv( np.matmul(X.T,X) ), np.matmul(X.T,Y))
print(theta)
```

```python
y_pred=np.matmul(X,theta)
print(y_pred)
```
